- $f(x_1,x_2)=\frac{1}{2} x_1^2 + \frac{9}{2} x_2^2$ at $x=(0,0)^T$.


```python
import numpy as np

def func1(x):
    f = 0.5*x.item(0)**2 + 4.5*x.item(1)**2
    g = np.array([[x.item(0)],[9*x.item(1)]])
    return (f,g)

x1 = np.array([0,0]).T
(f1,g1) = func1(x1)
print('The value of the function at x=',x1,' is f(x)=',f1,', \n and its gradient is grad(f)(x)=',g1.T)
```

```python
import matplotlib.pyplot as plt

d1x1 = - g1
d2 = np.array([[1],[1]])
d3 = np.array([[-2],[3]])


plotalpha = np.arange(0,1,0.01)
fig,ax = plt.subplots()
fx1 = np.array([func1(x1 + alpha * d1x1)[0] for alpha in plotalpha])
fx2 = np.array([func1(x1 + alpha * d2)[0] for alpha in plotalpha])
fx3 = np.array([func1(x1 + alpha * d3)[0] for alpha in plotalpha])
plot1,plot2,plot3 = plt.plot(plotalpha,fx1,plotalpha,fx2,plotalpha,fx3)
ax.legend((plot1,plot2,plot3), ('$d_1=-\\nabla f(x)$', '$d_2$','$d_3$'), loc='lower left', shadow=True)
plt.show()
```
