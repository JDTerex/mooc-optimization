```python
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt
```

Consider $f:\mathbb{R}\to\mathbb{R}$ defined as \\[f(x) = -x^4+12x^3-47 x^2 + 60 x. \\]

```python
def theFunction(x):
    return(-x**4 + 12 * x**3 - 47 * x**2 + 60 * x)
x = np.arange(0, 5.5, 0.1)
y = theFunction(x)
plt.plot(x, y)
```

The first derivative is \\[f'(x) = - 4 x^3 + 36 x^2 - 94x + 60. \\]

```python
def fprime(x):
    return -4 * x**3 + 36 * x**2 - 94 * x + 60
```

The second derivative is \\[ f''(x) = -12 x^2 + 72 x - 94. \\]

```python
def fsecond(x):
    return -12 * x**2 + 72 * x - 94
```

The quadratic model at the point $\widehat{x}$ is defined as \\[
    m(x) = f(\widehat{x}) + (x-\widehat{x}) f'(\widehat{x}) + \frac{1}{2} (x-\widehat{x})^2 f''(\widehat{x}) \\]

```python
def model(x, xhat):
    return theFunction(xhat) + (x - xhat) * fprime(xhat) + 0.5 * (x - xhat)**2 * fsecond(xhat)
```

The first derivative of the quadratic model is zero at the point \\[x^+ = \widehat{x} - \frac{f'(\widehat{x})}{f''(\widehat{x})}. \\]

```python
def zero(xhat):
    return xhat - fprime(xhat) / fsecond(xhat)
```

Plot the quadratic model at a given point

```python
def plotModel(xhat):
    print(f'f({xhat})={theFunction(xhat)}')
    print(f'f\'({xhat})={fprime(xhat)}')
    print(f'f\'\'({xhat})={fsecond(xhat)}')
    x = np.arange(0, 5.5, 0.1)
    y1 = theFunction(x)
    y2 = model(x, xhat)
    plt.ylim(-10, 25)
    plt.plot(x, y1)
    plt.plot(x, y2)
    plt.plot(xhat, theFunction(xhat), 'go')
    z = zero(xhat)
    print(f'Zero of the derivative of the model: {z:.3g}')
    plt.plot(z, model(z, xhat), 'ro')
```

Quadratic model around $\widehat{x}=3$.

```python
plotModel(3)
```

The model is convex. Indeed, the second derivative is positive. The zero of the derivative is therefore the minimum of the quadratic model. And there is a good adequacy between the model and the function at that point. This is a favorable case for Newton's local method.

Quadratic model around $\widehat{x}=4$.

```python
plotModel(4)
```

The model is convex. Indeed, the second derivative is positive. The zero of the derivative is therefore the minimum of the quadratic model. However, there is not a good adequacy between the model and the function at that point. The value of the function actually increase at that point. This is not a favorable case for Newton's local method.

Quadratic model around $\widehat{x}=5$.

```python
plotModel(5)
```

The model is not convex. Indeed, the second derivative is negative. Therefore, the model is not bounded from below and there is no minimum. The zero of the derivative corresponds to a __maximum__ of the quadratic model. Even if there is  a good adequacy between the model and the function at that point, the value of the function actually increase. This is not a favorable case for Newton's local method.
